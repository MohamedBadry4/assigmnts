### 7-Find Common Elements
### Write a Python function that takes two lists as input and returns a new list containing the common elements present in both lists.

```python
def elemnts(list1,list2):
  result = []
  for elemnt in list1 :
    if elemnt in list2:
       result.append(elemnt)
  return result
list1 = [1,5,7,8,9]
list2 = [1,2,5,6,7,8,9,10]
print(elemnts(list1,list2))
```

```python
def is_prime(number):

    if number <= 1:
        return False
    elif number == 2:
        return True
    elif number % 2 == 0:
        return False
    for i in range(3, int(number ** 0.5) + 1, 2):
        if number % i == 0:
            return False

    return True
# Example usage:
number_to_check = 57
if is_prime(number_to_check):
    print(f"{number_to_check} is a prime number.")
else:
    print(f"{number_to_check} is not a prime number.")
```

### 17-write a Python program to swap first and last element of any list.

```python
def swaplist(list1):
   size = len(list1)

   temp = list1[0]
   list1[0] = list1[size-1]
   list1[size-1] = temp
   return list1
list1 = [1,5,7,89,11,3]
print(swaplist(list1))


```

### 21-Write a Python program to drop empty items from a given dictionary.Original Dictionary:{'c1': 'Red', 'c2': 'Green', 'c3': None}New Dictionary after dropping empty items:{'c1': 'Red', 'c2': 'Green'}

```python
dict1 = {'c1': 'Red', 'c2': 'Green', 'c3':None}
print("Original Dictionary:")
print(dict1)
print("New Dictionary after dropping empty items:")
dict1 = {key:value for (key, value) in dict1.items() if value is not None}
print(dict1)
```
