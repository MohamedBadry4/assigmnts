```python

class Book:

    def __init__(self, title, author, genre, year):
        self.title = title
        self.author = author
        self.genre = genre
        self.year = year

    def __str__(self):
        return f"{self.title} by {self.author} ({self.year}) - {self.genre}"

class Library:

    def __init__(self):
        self.books = {}

    def add_book(self, book):
        if book.title in self.books:
            print("This book already exists in the library.")
        else:
            self.books[book.title] = book
            print("Book added successfully.")

    def remove_book(self, title):
        if title in self.books:
            del self.books[title]
            print("Book removed successfully.")
        else:
            print("This book does not exist in the library.")

    def search_book(self, query):
        matches = []
        for book in self.books.values():
            if query.lower() in book.title.lower() or query.lower() in book.author.lower():
                matches.append(book)
        return matches
    def display_books(self):
        if len(self.books) == 0:
            print("The library is empty.")
        else:
            for book in self.books.values():
                print(book)

    def save_data(self, filename):
        with open(filename, "w") as file:
            for book in self.books.values():
                file.write(f"{book.title},{book.author},{book.genre},{book.year}\n")
            print("Library data saved to file.")

    def load_data(self, filename):
        try:
            with open(filename, "r") as file:
                for line in file:
                    data = [item.strip() for item in line.split(",")]
                    book = Book(data[0], data[1], data[2], data[3])

                    self.books[book.title] = book
                print("Library data loaded from file.")
        except FileNotFoundError:
            print("File not found.")

library = Library()

# Load the library data from a file (if exists)
library.load_data("library.txt")
# ...

while True:
    print("\nWelcome to your personal library!")
    print("Please choose an option:")
    print("1. Add a book")
    print("2. Remove a book")
    print("3. Search for a book")
    print("4. Display all books")
    print("5. Save and exit")

    try:
        choice = int(input("> "))
        if choice < 1 or choice > 5:
            raise ValueError

        if choice == 1:
            title = input("Enter the title of the book: ")
            author = input("Enter the author of the book: ")
            genre = input("Enter the genre of the book: ")
            year = input("Enter the publication year of the book: ")

            if not title or not author or not genre or not year:
                print("Invalid input. All fields are required.")
            else:
                new_book = Book(title, author, genre, year)
                library.add_book(new_book)

        elif choice == 2:
            title = input("Enter the title of the book you want to remove: ")
            library.remove_book(title)

        elif choice == 3:
            query = input("Enter the title or author to search for: ")
            matches = library.search_book(query)
            if not matches:
                print("No matches found.")
            else:
                print("Matching books:")
                for match in matches:
                    print(match)

        elif choice == 4:
            library.display_books()

        elif choice == 5:
            library.save_data("library.txt")
            print("Library data saved. Goodbye!")
            break

    except ValueError:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
